```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/home/sonn/Sonn/Workspace/Projects/Anti-Obesity_GBC_NetworkPharmacology/data/final/BindingEnergies.csv")
```

```python
data.set_index(data.columns[0], inplace=True)
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(data, annot=True, cmap="coolwarm", fmt=".2f", cbar_kws={'label': 'Binding Energy (kcal/mol)'})
plt.ylabel("Compounds")
plt.xlabel("Targets")
plt.show()
```
