```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
BO_final = BO_final.sort_values(by='Count', ascending=False)
```

```python
CC_final = CC_final.sort_values(by='Count', ascending=False)
```

```python
plt.figure(figsize=(18, 10))
sns.barplot(y='Count', x='BO', data=BO_final, palette="viridis")

plt.ylabel('Enrichment score', fontsize=12)
plt.xlabel('Biological Process (BP)', fontsize=12)
plt.xticks(fontsize=12, rotation=90)
plt.yticks(fontsize=12)


plt.figure(figsize=(18, 10))
sns.barplot(y='Count', x='CC', data=CC_final, palette="viridis")

plt.ylabel('Enrichment score', fontsize=12)
plt.xlabel('Cellular Component (CC)', fontsize=12)
plt.xticks(fontsize=12, rotation=90)
plt.yticks(fontsize=12)


plt.figure(figsize=(18, 10))
sns.barplot(y='Count', x='MF', data=MF_final, palette="viridis")

plt.ylabel('Enrichment score', fontsize=12)
plt.xlabel('Molecular Function (MF)', fontsize=12)
plt.xticks(fontsize=12, rotation=90)
plt.yticks(fontsize=12)

# Hiá»ƒn thá»‹
plt.tight_layout()
plt.show()
```
